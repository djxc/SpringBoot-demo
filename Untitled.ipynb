```python
import matplotlib.pyplot as plt  
import numpy as np 
import math

fb = 50/60
fs = 6.6/60
a = 2.1/60
infil = []
for i in range(200):
    f = fs + (fb - fs)*math.exp(-a*i)
    infil.append(f)
print(infil)
```

```python
ii = []
for i in range(200):
    ii.append(i)

plt.plot(ii, infil, color='red')
plt.show()
```

```python
print(fb, fs, a)
```

```python
infil2 = []
for i in range(300):
    f = fs * i + ((fb - fs)*(1 - math.exp(-a*i)))/a
    infil2.append(f)
plt.plot(ii, infil2, color='red')
plt.show()
```
